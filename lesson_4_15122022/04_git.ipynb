### Verzování s Git-em

---

**Verzování** (obecně *VCS*), znamená **nahrávání** a **sdílení** vašich zdrojových souborů v rámci vzdálených uložišť.

<br>

Určitě by stačilo nahrávat řešení někam na cloud (Drive, Dropbox, Mega, aj.). Ale existují šikovnější řešení.

<br>

**Git** je nástroj, který vám umožňuje sdílet verzi vašeho zápisu s vašimi spolupracovníky a kolegy.

<br>

#### Nachystáme GitHub

---

Existuje několik řešení, my si povíme o [GitHub](https://github.com/about) (GitLab, Bitbucket, Beanstalk)

<br>

Předně jaký je vztah mezi **Git** a **GitHub**. Velice jednodušše:

<br>

**Git** je obecně softwarové řešení pro nahrávání, sdílení a evidování změn.

<br>

**GitHub** je v podstatě implementace Gitu, která umožňuje interakce se vzdáleným uložištěm.

<br>

#### Celý postup

---

<br>

<br>

#### Vytvoříme **nový účet** na [GitHub](https://github.com/)

jakmile máš účet *vytvořený*, můžeš se **přihlásit**.

<img src="https://i.imgur.com/vFhqhW8.png" width="600">

<br>

#### Vytvoříme **nové uložiště** (repozitář, nebo také archív příp. digitální úschovna),
v pravém horním rohu obrazovky (vedle tlačítka svého účtu) vybereš symbol `+` a označíš možnost nový repozitář.

<img src="https://i.imgur.com/o1PXrVF.png" width="300">

#### Zadáme několik **detailů**
určitě potřebuješ:
1. zadat **jméno projektu** (např. *muj_prvni_projekt*),
2. nastavit projekt jako **veřejný** repozitář (aby jej lektor mohl otevřít),
3. kliknout na tlačítko `Create repository`.

<img src="https://i.imgur.com/wfdtGiY.png" width="600">

Následně klikneš na variantu `HTTPS` a potom zkopíruješ **odkaz** (tohoto nově vytvořeného repozitáře, pomocí tlačítka v pravo.

<img src="https://i.imgur.com/fzKs1bq.png" width="600">

<br>

#### Propojíme VS code a GitHub
- vše přes tlačítko vlevo Source control --> a poté tři tečky nahoře 
- clone pro stažení repozitáře z Githubu (přes tři tečky)
- stage changes (tlačítko "+" u vybraného souboru) pro přidání změn do indexu
- commit pro uložení změn (přes tři tečky)
- push pro poslání změn do Githubu přes tři tečky

#### Podepíšeš se pro přidávání příspěvků
otevřeš v PyCharmu kartu `Terminal` a zapíšeš příkazy:
```bash
git config user.name "Moje jmeno"
git config user.email "Muj email"
```

<img src="https://i.imgur.com/ruRs0ue.gif" width="600">

#### Přidáš příspěvek a odesíláš na vzdálený repozitář
Postupně:
1. Vybereš soubor, který chceš odeslat,
2. napíšeš příspěvek (commit),
3. Klikneš na tlačítko `commit and push` (přidat příspěvěk a odeslat).

<img src="https://i.imgur.com/87H2gi3.png" width="300">

Pokud na tebe vyskočí okno, které vyžaduje přihlášení, klikni na možnost `LOG IN VIA GITHUB...`

<img src="https://i.imgur.com/387twUv.png" width="400">

následně na tebe vyskočí rekapitulační okno, kde klikneš na modré tlačítko `Push`.

Potom už můžeš zkontrolovat svůj účet v prohlížeči, kde uvidíš tebou nahraný soubor (případně více souborů).
