### Komunikace s API, funkce `post`

---
Web API = application programming interface, slouží ke komunikaci se serverem, požadavky jsou na server odeslány ve formátu XML nebo JSON.

Funkce `post` umožňuje odeslat na server konkrétní dotaz, např. na získání výsledů vehledávání, výpis a úpravu databáze a podobně.
